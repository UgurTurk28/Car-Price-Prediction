```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
DF = pd.read_csv("merc.csv")
DF
```

```python
sbn.distplot(DF["price"])
```

```python
DF.sort_values("price", ascending = False)
```

```python
newdf = DF.sort_values("price", ascending = False).iloc[131:]
newdf
```

```python
newdf = newdf[DF.year != 1970]
newdf
```

```python
newdf = newdf.drop(columns = ["fuelType"])
newdf
```

```python
newdf = newdf.drop(columns = ["transmission"])
newdf = newdf.drop("model", axis=1)
newdf
```

```python
y = newdf["price"].values
x = newdf.drop("price", axis=1).values 
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(12, activation = "relu"))
model.add(Dense(12, activation = "relu"))
model.add(Dense(12, activation = "relu"))
model.add(Dense(12, activation = "relu"))

model.add(Dense(1))

model.compile(optimizer="adam", loss="mse")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test), batch_size=250, epochs = 300)
```

```python
kayipverisi = pd.DataFrame(model.history.history)
```

```python
testtahminleri = model.predict(x_test)
testtahminleri = pd.Series(testtahminleri.reshape(3897,))
testtahminleri
```

```python
tahmindf = pd.DataFrame (y_test, columns=["Gerçek fiyat"])
tahmindf
```

```python
tahmindf = pd.concat([tahmindf,testtahminleri], axis=1)
tahmindf
```
